###### 2단계 : 최고기온이 높은 날의 날짜와 그 날의 기온 값 찾기

```python
max_temp = -999
max_data = ''

for row in data:
    if max_temp < row[-2]:
        max_temp = row[-2]
        max_data = row[0]
```

###### 1단계 : 내가 태어난 일자의 평균 기온을 리스트로 만들기 (내 생일 2/14)

```python
result = []
for row in data:
    if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14':
        result.append(float(row[1]))
```

```python
result = []
for row in data:
    if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14':
        if row[1] != '':
            result.append(float(row[1]))
print(result)
```

```python
import matplotlib.pyplot as plt

plt.title('내 생일 평균 기온 시각화')
plt.plot(result, color='red', label='평균기온')
plt.legend()
plt.show()
```

```python
sum = 0
for i in result:
    sum+=i

print(f"기상이 관측된 {len(result)}해 동안, 내 생일의 평균 기온은 {sum/len(result) : .2f}이다.")
```

```python
christmas = []

for row in data:
    if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '25':
        if row[2] != '' and row[3] != '':
            christmas.append(float(row[2]) - float(row[3]))
```

```python
import matplotlib.pyplot as plt

plt.title('크리스마스 일교차 시각화')
plt.plot(christmas, color='green', label='크리스마스 일교차')
plt.legend()
plt.show()
```

```python
print(christmas)
```

```python
sum = 0
for i in christmas:
    sum+=i
print(f"크리스마스의 일교차 평균은 {sum/len(christmas) : .2f}이다.")
```
