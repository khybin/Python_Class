```python
sum = 0
count = 0
avg_age = 0
for row in data:
    if row[5] != '':
        sum += float(row[5])
        count += 1

avg_age = sum/count
print("평균 나이는", int(avg_age))
```

```python
for row in data:
    if row[5] == '':
        row[5] = int(avg_age)
        
print(data[5])
```

```python
Embarked = []
for row in data:
    if row[11] != '':
        Embarked.append(row[11])

print(Embarked)
```

```python
for row in data:
    if row[11] == '':
        row[11] = 'S'
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(dpi=80)
plt.title('타이타닉 탑승향 빈도수')
plt.hist(Embarked, label='Embarked(S, C, Q)')
plt.show()
```

```python
survived = {'0':0, '1':0}

for row in data:
    if row[1] == '0':
        survived['0'] += 1
    elif row[1] == '1':
        survived['1'] += 1
        
print(survived)

plt.figure(dpi=80)
plt.title('타이타닉 사망자 생존자 비율 세로막대 그래프')
plt.bar(survived.keys(),survived.values())
plt.show()
```

```python
SBG = {'생존 남성' : 0, '사망 남성' : 0, '생존 여성' : 0, '사망 여성' : 0}

for row in data:
    if row[4] == 'male':
        if row[1] == '0':
            SBG['사망 남성'] +=1
        elif row[1] == '1':
            SBG['생존 남성'] +=1
    else :
        if row[1] == '0':
            SBG['사망 여성'] +=1
        elif row[1] == '1':
            SBG['생존 여성'] +=1

print(SBG)
```

```python
plt.figure(figsize=(7, 3))
plt.bar(SBG.keys(), SBG.values(), color=['b', 'orange', 'b', 'orange'])
plt.show()
```

```python
plt.figure(figsize=(9, 4))
plt.pie(SBG.values(), labels=SBG.keys(),autopct='%.1f%%')
plt.show()
```

```python
Ticket = ['1st Sur', '1st Dead', '2nd Sur', '2nd Dead', '3rd Sur', '3rd Dead']
corre = [0,0,0,0,0,0]

for row in data:
    if row[2]=='1':
        if row[1] == '1':
            corre[0]+=1
        else:
            corre[1]+=1
    elif row[2] == '2':
        if row[1] == '1':
            corre[2] +=1
        else:
            corre[3] +=1
    else:
        if row[1] == '1':
            corre[4]+=1
        else:
            corre[5]+=1
            
print(corre)
```

```python
plt.bar(Ticket, corre)
plt.show()
```
